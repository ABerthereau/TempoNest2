#Example 2 - Tutorial

This Notebook explains a basic analysis analysis of the Example 2 data set. This data set uses a complete binary timing model, and both DM and scattering variations, with a simple Gaussian profile that does not evolve across the band.  As in Example 1, the analysis steps we take will be:

    1. Load the data (par/tim/profiles)
    2. 'Fold' the whole data set in time in order to form an average profile
    3. Fit an initial guess to the average profile using this time averaged data set.
    4. Get an initial estimate of the overall phase offset
    5. Use these initial estimates to form the Hessian of the likelihood (required by the sampler)
    6. Perform some burnin
    7. Load the chains, and use the new maixmum likelihood point from that analysis to redefine the Hessian.
    8. Perform the actual sampling
    9. Check the results.


# Load and Time Average The Profile Data

In this example we will average the profile data down into 4 channels so that we can get an initial estimate of the scattering timescale.  We load the par and tim file, and perform 4 iterations within tempo2, and then use the final timing model fit to peform the averaging.  We then output the scrunched data as a pickle object.

```python
import numpy as np
from Class import *

lfunc = Likelihood(useGPU = False)
lfunc.loadPulsar("Example2.par", "Example2.tim", root='./results/Example2-', usePreFit=False, iters=4)

lfunc.TScrunch(doplot = True, channels = 4, ToPickle = True, FromTM = True)
```

#Get Initial Profile Model

We can now perform an analysis of the scrunched data to get an initial estimate of the mean profile model. We now specify that we would like to include a scattering timescale in the model.



```python
lfunc.getInitialParams(pmin = [-0.5, -1.3010299956639813, 0, -10], pmax = [0.5, -1.3010299956639813, 0.1, 0], resume=True, outDir = './InitFFTMNChains/Example2-', sampler='multinest', incScattering = True, mn_live = 1000,  fitNComps = 1, doplot = False)
```

#Produce interpolation matrix and obtain intial phase estimate

```python
'''Make interpolation Matrix'''
lfunc.PreComputeFFTShapelets(interpTime = 1, MeanBeta = lfunc.MeanBeta, doplot=False)


lfunc.getInitialPhase(doplot = False, ToPickle = True)
lfunc.ShiftSats(1-lfunc.MeanPhase)
lfunc.MeanPhase = 0
```

#Setup The Model and Perform Burnin

Parameters are added to the model using the "addX" functions, where 'X" is the parameter in question. Each parameter can be passed a maximum likelihood set of values, otherwise they are determined in some fashion automatically. Each parameter can also be fit for, or fixed at the values provided/calculated.

In this case we add :

    * The instrument noise in the profile data,
    * The amplitudes of the individual profiles,
    * The phase offset,
    * The linear timing model,
    * The profile model.
    * The scattering variations model

We use the initial estimate of scattering from the previous stage to set the start point for the burnin for these parameters.  We then allow it to collect 10000 samples with a small step size to navigate to the overall maximum.

lfunc.ScatterCrossTerms specifies the amount of cross correlation to include in the hessian for the scattering variables.  As we are likely far from the overall maximum, we set this to be zero for burnin as the hessian will not be numerically stable.



```python

lfunc.DenseParams = 0	
lfunc.DiagParams = 0
lfunc.LinearParams = 0
lfunc.MLParameters = []
lfunc.parameters = []
lfunc.ParametersToWrite = []
lfunc.ParamDict = {}

lfunc.addPNoise(Fit = True, write=False)
lfunc.addPAmps(Fit = True, Dense=False, write=False)
lfunc.addPhase(Fit = True, ML=np.array([0]))
lfunc.addLinearTM(Fit = True)
lfunc.addProfile(Fit = True, ML = lfunc.MLShapeCoeff[1:].flatten())

MLScatter = np.ones(100)*lfunc.MeanScatter 
lfunc.addScatter(FitScatter = True, FitFreqScale = False, MLScatter = MLScatter, MLFreqScale = 4, RefFreq=1.4)
lfunc.ScatterCrossTerms = np.ones(lfunc.NScatterEpochs)*0


lfunc.n_params=len(lfunc.parameters)
lfunc.PhasePrior = 1e-1
lfunc.startPoint, lfunc.EigV, lfunc.EigM, lfunc.hess = lfunc.calculateGHSHessian(diagonalGHS=False)

lfunc.root='./results/Example2-Burnin-'
if(lfunc.DiagParams < lfunc.n_params):
	DenseParams = lfunc.startPoint[lfunc.DiagParams:]
	PrincipleParams = np.dot(lfunc.EigM.T, DenseParams)
	lfunc.startPoint[lfunc.DiagParams:] = PrincipleParams
	
lfunc.callGHS(resume=False, nburn = 100, nsamp = 10000, feedback_int = 100, seed = -1, max_steps = 10, dim_scale_fact = 0.04)
```

#Perform the Actual Sampling

We now perform the actual samping. As we will be collecting far more points we no longer write the profile noise/amplitudes to file. We also set 'dense = True' for the profile amplitudes, so they are included in the dense part of the hessian.  We also set lfunc.ScatterCrossTerms to be one. 

Finally we have now included the frequency scaling of the scattering as a free parameter aswell.


```python
chains=np.loadtxt('./results/Example2-Burnin-extract.dat').T
ML = chains.T[-100:][np.argmax(chains[-1][-100:])][:-1]


lfunc.DenseParams = 0	
lfunc.DiagParams = 0
lfunc.LinearParams = 0
lfunc.MLParameters = []
lfunc.parameters = []
lfunc.ParametersToWrite = []
lfunc.ParamDict = {}

lfunc.addPNoise(Fit = True, write=False)
lfunc.addPAmps(Fit = True, Dense=True, write=False)
lfunc.addPhase(Fit = True, ML=np.array([ML[0]]))
lfunc.addLinearTM(Fit = True, ML=ML[1:1+lfunc.numTime])
lfunc.addProfile(Fit = True, ML = lfunc.MLShapeCoeff[1:].flatten())


lfunc.addScatter(FitScatter = True, FitFreqScale = True, MLScatter = ML[-100:], MLFreqScale = 4, RefFreq=1.4)
lfunc.ScatterCrossTerms = np.ones(lfunc.NScatterEpochs)

lfunc.n_params=len(lfunc.parameters)
lfunc.PhasePrior = 1e-1
lfunc.startPoint, lfunc.EigV, lfunc.EigM, lfunc.hess = lfunc.calculateGHSHessian(diagonalGHS=False)



lfunc.root='./results/Example2-Burnt-'
if(lfunc.DiagParams < lfunc.n_params):
	DenseParams = lfunc.startPoint[lfunc.DiagParams:]
	PrincipleParams = np.dot(lfunc.EigM.T, DenseParams)
	lfunc.startPoint[lfunc.DiagParams:] = PrincipleParams
	
	
#lfunc.callGHS(resume=False, nburn = 100, nsamp = 20000, feedback_int = 100, seed = -1, max_steps = 10, dim_scale_fact = 0.4)
```

```python
Dvals=np.loadtxt("./results/Example2-Burnt-DMVals.dat")
SimDvals = np.loadtxt("./results/SimDMVals.dat")

plt.errorbar(Dvals[:,0], Dvals[:,1], yerr=Dvals[:,2], linestyle='None', color='red')
plt.scatter(Dvals[:,0], SimDvals, color='black')
plt.show()

Diff=Dvals[:,1]-SimDvals
Diff = (Diff - np.mean(Diff))/Dvals[:,2]
bins=np.linspace(-4,4,14)
xbins=(bins[:-1]+bins[1:])/2
gauss=np.exp(-0.5*xbins**2)
gauss = (gauss/np.sum(gauss))*99
hist,bins=np.histogram(Diff, bins=bins)
plt.plot(xbins, gauss, linestyle='steps')
plt.plot(xbins, hist, linestyle='steps')
plt.show()
```

```python
Svals=np.loadtxt("./results/Example2-Burnt-ScatterVals.dat")
SimSvals = np.loadtxt("./results/SimScatterVals.dat2")

MeanScatter = np.mean(SimSvals)

plt.errorbar(Svals[:,0], Svals[:,1]*1000-MeanScatter, yerr=Svals[:,2]*1000, linestyle='None', color='red')
plt.scatter(Svals[:,0], SimSvals-MeanScatter, color='black')
plt.show()



Diff=Svals[:,1]*1000-SimSvals
Diff = (Diff - np.mean(Diff))/(Svals[:,2]*1000)
bins=np.linspace(-4,4,14)
xbins=(bins[:-1]+bins[1:])/2
gauss=np.exp(-0.5*xbins**2)
gauss = (gauss/np.sum(gauss))*99
hist,bins=np.histogram(Diff, bins=bins)
plt.plot(xbins, gauss, linestyle='steps')
plt.plot(xbins, hist, linestyle='steps')
plt.show()
```
